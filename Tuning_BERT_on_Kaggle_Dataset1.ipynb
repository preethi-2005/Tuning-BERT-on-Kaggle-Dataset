## Step 1: Create Google Gemini API Key

To access Gemini models, you must create an API key.

### Procedure:

1. Go to **Google AI Studio**
   Google AI Studio

2. Click on **“Get API Key”**

3. Click **“Create API Key”**

4. Choose or create a project (e.g., “demo”)

5. Generate the key

6. Copy the key immediately

7. Paste it into a text file (for example):

   ```
   gemini_key.txt
   ```

Important:

* Do not leave spaces before or after the key.
* Save it properly.
* Keep it secure.

Google Colab

## Step 3: Read the API Key in Python

Conceptual structure:

```python
with open("gemini_key.txt", "r") as f:
    api_key = f.read().strip()
```

* `"r"` → Read mode
* `strip()` → Removes unwanted spaces or newline characters

Now your API key is stored in the variable `api_key`.

## Alternative Method: Using Chat-Based Interface

Instead of:

```python
client.models.generate_content()
```

You can use chat-style interaction.

Example structure:

```python
chat = client.chats.create(
    model="gemini-2.5-flash"
)

response = chat.send_message("Who is the Indian president?")
print(response.text)
```

### What Is Happening Here?

* `client.chats.create()` → Creates a chat session
* `send_message()` → Sends prompt dynamically
* `response.text` → Retrieves model response

This is similar to how chat applications work internally.

---

## Difference Between the Two Methods

### Method 1: generate_content()

* Direct prompt
* Single request-response
* Stateless

### Method 2: chats.create() + send_message()

* Conversational
* Maintains context (if multiple messages)
* Closer to chatbot behavior

Both are valid.

Use whichever fits your application design.

# Steps to create Grok Api keys

````python
from groq import Groq

with open("grok_api.txt", "r") as f:
    grok_api_key = f.read().strip()

client = Groq(api_key=grok_api_key)
````

````python
response = client.chat.completions.create(
    model="llama-3.1-8b-instant",
    messages=[
        {"role": "user", "content": "Who is the Indian president?"}
    ]
)

print(response.choices[0].message.content)
````

````python
response = client.chat.completions.create(
    model="llama-3.1-8b-instant",
    messages=[{"role": "user", "content": "Explain transformers"}]
)
print(response.choices[0].message.content)
````
